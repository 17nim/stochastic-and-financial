```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
n = 10000
p = 0.3
N = 10
```

```python
X = np.random.binomial(N, p, n)
```

```python
minX = X.min()
```

```python
maxX = X.max()
```

```python
e = np.arange(minX-0.5, maxX+1)
```

```python
plt.hist(X, bins=e)
plt.title("Histogram with bins = e")
plt.show()
```

```python
Mn = np.mean(X)
```

```python
pn = Mn / N
```

```python
print("Mean of X", Mn)
```

```python
print("Probability pn", pn)
```

```python
from scipy import stats
```

```python
from scipy.stats import binom
```

```python
k = np.arange(0, N+1)
```

```python
PMF = binom.pmf(k, N, pn)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(k, PMF, 'ro', ms=12, mec='r')
ax.vlines(k, 0, PMF, colors='r', lw=4)
plt.show()
```

```python
H = list(np.bincount(X))
for j in range(len(H), N+1):
    H.append(0)
```

```python
Z = 0
for j in range(N + 1):
    Z += (H[j] - n * PMF[j]) ** 2 / (n * PMF[j])
Z
```
